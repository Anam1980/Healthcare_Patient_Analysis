```python
import pandas as pd

df = pd.read_csv("healthcare_patient_data_cleaned.csv")
```

```python
# Age Group
df["age_group"] = pd.cut(df['age'],
                         bins=[0, 30, 45, 60, 100],
                        labels=['Young', 'Adult', 'Middle_Aged', 'Senior'])
```

```python
# High/Low Blood Pressure Flag
df['high_bp_flag'] = df["blood_pressure"].apply(lambda x: "High_BP" if x >= 140 else "OK")
```

```python
# High Blood Glucose Flag (Diabetes Indicator)
df['high_glucose_flag'] = df['blood_glucose'].apply(lambda x: "Diabetes" if x>=180 else "OK")
```

```python
# Comorbidity Risk Score
df['risk_score'] = df['comorbidity_count'] + df['prior_admissions_6mo']
```

```python
# Length of Stay Category
df['stay_type'] = pd.cut(
    df['length_of_stay'],
    bins=[0, 3, 7, 15],
    labels=['Short', 'Medium', 'Long']
)
```

```python
# Total Medication Burden Feature
df['high_medication_burden'] = df['med_count'].apply(
    lambda x: "High" if x >= 7 else "OK"
)
```

```python
df = pd.get_dummies(df, columns=['gender', 'department', 'age_group', 'stay_type'], drop_first=True)
```

```python
df.to_csv("healthcare_feature_engineered.csv", index=False)
```
