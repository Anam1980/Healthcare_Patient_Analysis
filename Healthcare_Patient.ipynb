```python
df.drop_duplicates(inplace=True)
```

```python
df["admit_date"] = pd.to_datetime(df["admit_date"])
```

```python
df.info()
```

```python
df['admit_month'] = df['admit_date'].dt.month
df['admit_day'] = df['admit_date'].dt.day
df['admit_day_name'] = df['admit_date'].dt.day_name()
```

```python
df.info()
```

```python
df['gender'] = df['gender'].str.title()
df['department'] = df['department'].str.title()
```

```python
q1 = df['blood_glucose'].quantile(0.25)
q3 = df['blood_glucose'].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr

df['blood_glucose'] = df['blood_glucose'].clip(lower, upper)
```

```python
q1 = df['blood_pressure'].quantile(0.25)
q3 = df['blood_pressure'].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr

df['blood_pressure'] = df['blood_pressure'].clip(lower, upper)
```

```python
df.to_csv("healthcare_patient_data_cleaned.csv", index=False)
```
